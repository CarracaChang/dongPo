## [4] 匯入與 4-bit 量化設定（T4 自動用 float16）

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
from google import genai
from google.genai import types

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("裝置：", DEVICE)
if DEVICE == "cuda":
    name = torch.cuda.get_device_name(0)
    major, _ = torch.cuda.get_device_capability()
    print("GPU：", name)
    # T4 (capability 7.5) 不支援 bfloat16 → 用 float16；3070/A100 等 (>=8) 用 bfloat16
    COMPUTE_DTYPE = torch.bfloat16 if major >= 8 else torch.float16
    print("compute_dtype：", "bfloat16" if COMPUTE_DTYPE == torch.bfloat16 else "float16")
else:
    COMPUTE_DTYPE = torch.float16
    print("⚠️ 沒抓到 GPU。請到『執行階段 → 變更執行階段類型』選 T4 GPU。")

BNB_CONFIG = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=COMPUTE_DTYPE,
)
print("BNB 量化設定就緒")
```

## [5] 初始化 Gemini（史實查核模組）

```python
GEMINI_SYSTEM = (
    "你是一個專屬蘇東坡的歷史背景知識庫。請提供客觀、簡明的史實。"
    "【嚴格禁止】自稱人工智慧或AI。"
    "若使用者詢問蘇東坡的第一人稱問題（如：你喜歡吃什麼？），請直接輸出蘇東坡的歷史生平事實"
    "（如：蘇軾在黃州喜歡吃豬肉、竹筍等）。"
    "若使用者詢問後世歷史，請直接給出該歷史事件的客觀簡介。"
)

gclient = genai.Client(api_key=GEMINI_API_KEY)
GEMINI_MODEL_NAME = "gemini-2.5-flash"

def get_historical_facts(user_question: str) -> str:
    """用 Gemini 查客觀史實。新版 SDK 寫法，含簡單錯誤處理。"""
    try:
        resp = gclient.models.generate_content(
            model=GEMINI_MODEL_NAME,
            contents=user_question,
            config=types.GenerateContentConfig(system_instruction=GEMINI_SYSTEM),
        )
        return (resp.text or "").strip()
    except Exception as exc:
        msg = str(exc)
        if "429" in msg or "RESOURCE_EXHAUSTED" in msg:
            return "（史實查詢觸發流量限制 429，請稍後再試，或於 Secrets 換一把金鑰）"
        return f"（史實查詢失敗：{msg}）"

# 小測試（會真的呼叫一次 API，確認金鑰可用）
print("Gemini 測試回應：")
print(get_historical_facts("蘇軾在黃州的飲食"))
```

## [6] 載入東坡底模 + 融合 LoRA

```python
print("載入底模中…（第一次較久）")
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL_PATH, trust_remote_code=True)
base_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL_PATH,
    quantization_config=BNB_CONFIG,
    device_map="auto",
    trust_remote_code=True,
)
print("融合 LoRA 中…")
model = PeftModel.from_pretrained(base_model, LORA_WEIGHT_PATH)
model.eval()
print("✅ 東坡 + adapter 就緒")
if torch.cuda.is_available():
    print(f"顯存佔用：{torch.cuda.memory_allocated()/1024**3:.2f} GB")
```

## [7] 確認對話格式（重要）
東坡用的是**非標準** Human/Assistant 格式（不是標準 ChatML）。`apply_chat_template` 會讀東坡 tokenizer 自帶的 template。
這格把實際組出來的提示詞印出來，**你親眼確認格式對不對**，再往下用。

```python
_demo_msgs = [
    {"role": "system", "content": "（測試 system）"},
    {"role": "user", "content": "（測試 user 問題）"},
]
_demo = tokenizer.apply_chat_template(_demo_msgs, tokenize=False, add_generation_prompt=True)
print("=== apply_chat_template 實際輸出 ===")
print(_demo)
print("=== 檢查 ===")
print("含 <|im_start|>（標準ChatML特徵）：", "<|im_start|>" in _demo)
print("含 Human:/Assistant:（東坡自訂格式特徵）：", ("Human:" in _demo or "Assistant:" in _demo))
print("\n若是 Human/Assistant 格式即與底模一致，可放心使用。")
```

## [8] 組裝提示詞 + 本地推論

```python
SUDONGPO_SYSTEM = (
    "你現在是宋代文人蘇軾（蘇東坡）"
    "一律使用繁體中文，融合文言與白話，展現豁達幽默的個性。"
    "對於宋代之後的事物，勿直接以現代人視角回答。"
)

def assemble_prompt(historical_facts: str, user_question: str) -> str:
    return (
        "【史實參考】\n"
        f"{historical_facts}\n\n"
        "【處理指引】\n"
        "若使用者問的人事物發生在北宋（1101年）之後，你就說你不知道。"
        "【問題】\n"
        f"{user_question}\n"
    )

@torch.inference_mode()
def generate_response(assembled_prompt: str, max_new_tokens: int = 512, temperature: float = 0.7) -> str:
    messages = [
        {"role": "system", "content": SUDONGPO_SYSTEM},
        {"role": "user", "content": assembled_prompt},
    ]
    text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to(model.device)
    in_len = inputs["input_ids"].shape[1]
    out = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=temperature > 0,
        temperature=temperature if temperature > 0 else None,
        top_p=0.8,
        pad_token_id=tokenizer.pad_token_id or tokenizer.eos_token_id,
    )
    return tokenizer.decode(out[0][in_len:], skip_special_tokens=True).strip()

def ask_dongpo(user_question: str, show_facts: bool = True) -> str:
    facts = get_historical_facts(user_question)
    assembled = assemble_prompt(facts, user_question)
    answer = generate_response(assembled)
    if show_facts:
        print("="*60); print("【Gemini 史實】"); print(facts)
        print("-"*60); print("【蘇東坡回答】"); print(answer); print("="*60)
    return answer

print("推論函式就緒")
```

## [10]批次認知邊界測試（讀 Excel → 逐筆問 duo-LLM → 匯出）

```python
# =====================================================================
# [批次測試 1/3] 讀取 Excel 題庫 → 洗成 name 清單
# =====================================================================
import os
import pandas as pd

# 優先嘗試的檔案路徑（依你 Drive 實際位置調整第一個）
_CANDIDATES = [
    "/content/drive/MyDrive/FJU_dongpoProject/dongPo/ask/dongpo_edge_boundary_100_100.xlsx",
    "/content/drive/MyDrive/FJU_dongpoProject/dongPo/dongpo_knowledge_boundary.xlsx",
    "dongpo_edge_boundary.xlsx",
    "dongpo_knowledge_boundary.xlsx",
]
EXCEL_PATH = next((p for p in _CANDIDATES if os.path.isfile(p)), None)
assert EXCEL_PATH is not None, f"找不到 Excel，請確認路徑。已嘗試：{_CANDIDATES}"

df_src = pd.read_excel(EXCEL_PATH)

# 統一欄名：把 "name 名稱" / "name" 之類都對應到乾淨欄名
rename = {}
for col in df_src.columns:
    low = str(col).strip().lower()
    if low.startswith("name"):     rename[col] = "name"
    elif low.startswith("label"):  rename[col] = "label"
    elif low.startswith("type"):   rename[col] = "type"
    elif low.startswith("reason"): rename[col] = "reason"
df_src = df_src.rename(columns=rename)

assert "name" in df_src.columns, f"Excel 找不到 name 欄，實際欄位：{list(df_src.columns)}"

# 洗出 name：去空白、丟掉空列、去重、保持原順序
names = (
    df_src["name"].astype(str).str.strip()
    .replace("nan", "").replace("", pd.NA).dropna()
    .drop_duplicates().tolist()
)

print(f"來源檔：{EXCEL_PATH}")
print(f"洗出 name 數量：{len(names)}")
print("前 5 筆：", names[:5])
```

```python
# =====================================================================
# [批次測試 2/3] 逐一把 name 套進問句 → 丟給 Gemini+東坡 duo 架構
# =====================================================================
# 依賴前面已定義好的：get_historical_facts() / assemble_prompt() / generate_response()
import time

# 想先小量試跑就改這個上限；設 None 表示全部
LIMIT = None          # ← 確認沒問題後改成 None 跑全部
SLEEP_BETWEEN = 0.5 # 每題之間稍微停一下，降低 Gemini 觸發 429 機率

batch_results = []   # 收集結果，給第 3 格輸出用
_todo = names if LIMIT is None else names[:LIMIT]
print(f"本次將測試 {len(_todo)} 筆\n")

for i, nm in enumerate(_todo, 1):
    ask_text = f"先生，請問什麼是{nm}？"
    facts = get_historical_facts(ask_text)                 # Gemini 史實
    assembled = assemble_prompt(facts, ask_text)            # 組裝提示詞
    answer = generate_response(assembled)                   # 東坡回答

    batch_results.append({
        "name": nm,
        "ask": ask_text,
        "gemini_fact": facts,
        "system_ans": answer,
    })

    # 照指定格式輸出
    print("=" * 60)
    print(f"[{i}/{len(_todo)}] {nm}")
    print("【Gemini 史實】")
    print(facts)
    print("-" * 60)
    print("【蘇東坡回答】")
    print(answer)
    print("=" * 60 + "\n")

    time.sleep(SLEEP_BETWEEN)

print(f"完成，共收集 {len(batch_results)} 筆結果。")
```

```python
# =====================================================================
# [批次測試 3/3] 整理結果 → 存成 Excel → 下載
# =====================================================================
import datetime
import pandas as pd

assert batch_results, "batch_results 是空的，請先跑上一格的問答迴圈。"

df_out = pd.DataFrame(batch_results, columns=["name", "ask", "gemini_fact", "system_ans"])

ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
out_name = f"dongpo_duo_results_{ts}.xlsx"

# 存兩份：一份到 Drive（保險），一份到當前目錄（給瀏覽器下載）
drive_dir = "/content/drive/MyDrive/FJU_dongpoProject/dongPo/results"
saved_paths = []
try:
    import os
    os.makedirs(drive_dir, exist_ok=True)
    p1 = os.path.join(drive_dir, out_name)
    df_out.to_excel(p1, index=False)
    saved_paths.append(p1)
except Exception as e:
    print("（存到 Drive 失敗，略過）：", e)

df_out.to_excel(out_name, index=False)   # 當前目錄
saved_paths.append(out_name)

print("已存檔：")
for p in saved_paths:
    print("  ", p)
print(f"\n共 {len(df_out)} 筆。預覽前 3 筆：")
print(df_out.head(3).to_string())

# 觸發瀏覽器下載
try:
    from google.colab import files
    files.download(out_name)
except Exception as e:
    print("（自動下載失敗，可到 Drive 的 results 資料夾取檔）：", e)
```
